```python
def start(field,all_moves): # create game field
    for i in range(0, 9):
        field.append(str(i + 1))
        all_moves.append(i + 1)
    return

def print_square(field): # print game field after every move
    print('  0 1 2')
    vertical = -1
    for n in range(0, 9, 3):
        vertical += 1
        horizontal = ''
        for i in range(0, 3):
            horizontal += field[n + i] + " "
        print(vertical,horizontal)
    print('')

def p1_move(field, player_1, all_moves): # player_1 move
    print("Player_1, enter the nuber of cell:\n")
    x = int(input())
    while all_moves.count(x) == 0:
        print("Invalid numer of cell!")
        print("Player_1, enter the nuber of cell:\n")
        print(all_moves.count(x))
        print(all_moves)
        x = int(input())
    field[x-1] = player_1 # remove from possible moves
    all_moves.remove(x)
    return

def p2_move(field, player_2, all_moves): # player_2 move
    print("Player_2, enter the nuber of cell:\n")
    x = int(input())
    while all_moves.count(x) == 0:
        print("Invalid numer of cell!")
        print("Player_2, enter the nuber of cell:\n")
        print(all_moves.count(x))
        print(all_moves)
        x = int(input())
    field[x-1] = player_2 # remove from possible moves
    all_moves.remove(x)
    return

def check_horizontal(field, num): # check horizontal if win
    h = ''
    for i in range(0, 3):
        h += field[num*3 + i]
    return h

def check_vertical(field, num): # check verical if win
    v = ''
    for i in range(num, 9, 3):
        v += field[i]
    return v

def check_diag(field, num): # check diagonal if win
    if num == 1:
        return field[0]+field[4]+field[8]
    else: 
        return field[2]+field[4]+field[6]

def player_1_wins(player_1, win): # check if player_1 wins
    shablon = player_1*3
    if win.count(shablon)>0:
        print('Player_1 wins!')
        return True
    else:
        return False

def player_2_wins(player_2, win): # check if player_2 wins
    shablon = player_2*3
    if win.count(shablon)>0:
        print('Player_2 wins!')
        return True
    else:
        return False

def check(field, player_1, player_2): # check if someone wins
    winner = False
    win = []
    for i in range(0,3):
        win.append(check_vertical(field, i))
        win.append(check_horizontal(field, i))
    win.append(check_diag(field, 1))
    win.append(check_diag(field, 2))
    winner = player_1_wins(player_1, win) or player_2_wins(player_2, win)
    return winner

def main():
    print('Game beginns!')
    field = []
    all_moves = []
    start(field,all_moves)
    print_square(field)
    player_1, player_2 = 'X', "O" 
    i=1
    while not check(field, player_1, player_2):
        if len(all_moves) == 0:
            print('DRAW!')
            break
        if i == 1:
            i=2
            p1_move(field, player_1, all_moves)
        else:
            i=1
            p2_move(field, player_2, all_moves)
        print_square(field)
    continue_ = input("Would you like to play again? \n y or n \n")
    if continue_ == "y":
        main()
    else:
        print("Good bye!")

main()
```
